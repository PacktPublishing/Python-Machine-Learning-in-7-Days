## Predicting weight using linear regression

### Import the height-weight data csv

```python
import numpy
filename = "height-weight.csv"
raw_data = open(filename, 'rt')
data = numpy.loadtxt(raw_data, delimiter=",")
```

```python
# We check the size of the data
data.shape
```

```python
# We check the first 10 rows
data[:10]
```

```python
#We import the plotting library 
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
```

```python
#We separate out the independent variable height into X 
#and dependent variable weight into y
X=data[:,0]
y=data[:,1]
```

```python
X_20 = X[:20]
y_20 = y[:20]
plt.scatter(X_20, y_20, color='red', s=30) ## here s is size
```

```python
# Split the data into training/testing sets
X_train=X[:4500]
X_test=X[4500:]
# Split the targets into training/testing sets
y_train=y[:4500]
y_test=y[4500:]
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
# Modeify the data to input into sklearn
X_train=X_train.reshape(-1, 1)
X_test=X_test.reshape(-1, 1)
```

```python
X_train.shape
```

```python
# Create linear regression object
regr = linear_model.LinearRegression()
```

```python
# Train the model using the training sets
regr.fit(X_train, y_train)
```

```python
# Make predictions using the testing set
y_pred = regr.predict(X_test)
```

```python
# The coefficients
print('Coefficients: \n', regr.coef_)
```

```python
# Plot outputs
plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```
